```python
import sys
import grid
import pickle
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def my_plot(d = 2):
    res = []
    file_names = os.listdir('./'+ 'd = ' + str(d))
    for file_name in sorted(file_names):
        if file_name[-3:] == 'csv':
            data = pd.read_csv('./'+ 'd = ' + str(d)+'/'+file_name, index_col = 0)
            res.append(data)
    df = pd.concat(res, axis = 1)
    df.plot()
    plt.vlines([1/(2*d-1), 2/3], 0, 1, linestyles='dashed', colors='red')
    plt.show()
```

```python
my_plot(d = 2)
```

```python
my_plot(d = 3)
```
